# 从CSV读入时间序列数据
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
# 导入包
import pandas as pd

# CSV 文件名称
csv_file_name = 'SP500_2014-01-01_2022-12-31.csv'

# 读入CSV文件
df = pd.read_csv(csv_file_name)
```

```python
# 将输入的数据转换为日期时间对象
df["DATE"] = pd.to_datetime(df["DATE"])
```

```python
# 将名为"DATE"的列设置为索引
df.set_index('DATE', inplace = True)
```
